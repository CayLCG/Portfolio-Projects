```python
##############################################################
# Import the necessary packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

##############################################################
```

```python
##############################################################
# Import the CSV file from your computer using directory
# Shows the first 5 rows of the CSV file

Spotify_Tracks = pd.read_csv("C:/Users/dronc/Desktop/tracks.csv")
Spotify_Tracks.head()

##############################################################
```

```python
##############################################################
# Checking for null values
# Summing up all the null values under each category

pd.isnull(Spotify_Tracks).sum()

##############################################################
```

```python
##############################################################
# We see that every category has 71 more entries than names
# This is consistent with the 71 null values we detected
# We also see the number of different categories, datatypes in the CSV file as well as the storage space needed

Spotify_Tracks.info()

##############################################################
```

```python
##############################################################
# Sorting
# Shows the top 20 least popular songs

Least_Popular_Songs = Spotify_Tracks.sort_values('popularity', ascending = True).head(20)
Least_Popular_Songs

##############################################################
```

```python
##############################################################
# Sorting
# Shows the top 20 most popular songs

Most_Popular_Songs = Spotify_Tracks.sort_values('popularity', ascending = False).head(20)
Most_Popular_Songs

##############################################################
```

```python
##############################################################
# Shows the numerical measures using statistical terms

Spotify_Tracks.describe().transpose()

##############################################################
```

```python
##############################################################
# Shows the most popular songs with 80 and above popularity ranking

Above_80_Popularity = Spotify_Tracks.query('popularity >= 80', inplace = False).sort_values('popularity', ascending = False)
Above_80_Popularity

##############################################################
```

```python
##############################################################
# Sorting based on release date
Spotify_Tracks.set_index('release_date', inplace = True)
Spotify_Tracks.index = pd.to_datetime(Spotify_Tracks.index)
Spotify_Tracks.head()

##############################################################
```

```python
##############################################################
# Extracting the data from the 20th row

Spotify_Tracks[['artists']].iloc[20]

##############################################################
```

```python
##############################################################
# Converting the time from milliseconds to seconds

Spotify_Tracks["duration"] =  Spotify_Tracks['duration_ms'].apply(lambda t: round(t/1000))
Spotify_Tracks.drop('duration_ms', inplace = True, axis = 1)
Spotify_Tracks.duration.head()

##############################################################
```

```python
##############################################################
# Generates a heatmap to visualise the correlation between variables in the dataset. 
# Helps identify the strength and direction of linear relationships between different variables
# The higher the number, the higher the positive correlation

Correlation = Spotify_Tracks.drop(["mode", "key", "explicit"], axis = 1 ).corr(method = 'pearson')
plt.figure(figsize = (14, 6))
heatmap = sns.heatmap(Correlation,  annot = True, fmt = ".1g", vmin = 1, vmax = 1, center = 0, cmap = "rocket", linewidths = 1, linecolor = "Black")
heatmap.set_title("Correlation HeatMap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)

##############################################################
```

```python
##############################################################
# Filtering only some of the datasets for regression plot

Sample = Spotify_Tracks.sample(int(0.004 * len(Spotify_Tracks)))
print(len(Sample))

##############################################################
```

```python
##############################################################
# Shows the positive correlation between Loudness and Energy
# Consistent with the table showing the correlation between Loudness and Energy at value 1

plt.figure(figsize = (10, 6))
sns.regplot(data = Sample, y = 'loudness', x = 'energy', color = 'r').set(title = 'Correlation between Loudness and Energy')

##############################################################
```

```python
##############################################################
# Shows the negative correlation between Energy and Acousticness
# Consistent with the table showing the correlation between Energy and Acousticness at value -0.7

plt.figure(figsize = (10, 6))
sns.regplot(data = Sample, y = 'energy', x = 'acousticness', color = 'g').set(title = 'Correlation between Energy and Acousticness')

##############################################################
```

```python
##############################################################
# Shows the distribution of songs over the years using histogram

Spotify_Tracks['dates'] = Spotify_Tracks.index.get_level_values('release_date')
Spotify_Tracks.dates = pd.to_datetime(Spotify_Tracks.dates)
years = Spotify_Tracks.dates.dt.year
```

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = "hist", color = 'c').set(title = "Number of songs per year")

##############################################################
```

```python
##############################################################
# Shows the correlation between years and total duration of songs using histogram

Total_Duration = Spotify_Tracks.duration
fig, ax = plt.subplots(figsize = (18, 7))
fig = sns.barplot(x = years, y = Total_Duration, ax = ax, errwidth = False).set(title =  "Correlation between years and duration")
plt.xticks(rotation = 90)

##############################################################
```

```python
##############################################################
# Shows the correlation between years and total duration of songs using line plot

Total_Duration = Spotify_Tracks.duration
sns.set_style(style = "whitegrid")
fig, ax = plt.subplots(figsize = (10, 5))
fig = sns.lineplot(x = years, y = Total_Duration, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation = 65)

##############################################################
```
